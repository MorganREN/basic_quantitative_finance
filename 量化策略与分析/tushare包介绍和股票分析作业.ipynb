```python
import tushare as ts
import pandas as pd
import numpy as np
import matplotlib. pyplot as plt
```

### 使用tushare包获取某股票的历史行情数据

```python
pro = ts.pro_api()
df = pro.daily(ts_code='689009.SH')
df.to_csv("689009SH.csv")
```

```python
df = pd.read_csv("689009SH.csv", index_col='trade_date', parse_dates=['trade_date'])[['open','close', 'high','low']]
df
```

### 输出该股票所有收盘比开盘上涨3%以上的日期

```python
df[(df['close']-df['open'])/df['open'] >= 0.03].index
```

### 输出该股票所有开盘比前日收盘跌幅超过2%的日期

```python
df['close'].shift(-1) # shift使数据移动，括号内为正则下移，为负则上移
```

```python
df[(df['open']-df['close'].shift(-1))/df['close'].shift(-1) >= 0.02]
```

```python
df[(df['open']-df['close'].shift(-1))/df['close'].shift(-1) >= 0.02].index
```

### 加入我从2020年11月开始，每月第一个交易日买入1手股票，每年最后一个交易日卖出所有股票，到今天为止，我的收益如何？

```python
price_last = df['open'][0]
print(price_last)
df = df['2021-09-03':'2020-11'] # 剔除首尾无用数据
df
```

```python
df_monthly = df.resample("M").first() # 每月第一天的股票行情
df_monthly
```

```python
df_yearly = df.resample("A").last()[:1] # 每年的最后一天（不包含2021年）
df_yearly
```

```python
cost_money = 0 # 共花费钱数
hold = 0 # 共持有股数
for year in range(2020,2022):
    cost_money += df_monthly[str(year)]['open'].sum()*100 # 每一手买入买入100股所用钱数累计
    hold += len(df_monthly[str(year)]['open']) * 100 # 每一手买入买入100股
    
    if year != 2021:
        cost_money -= df_yearly.loc[str(year)]['open'][0] * hold
        hold = 0
    else: 
        continue
        
    print(cost_money)
```

```python
cost_money -= hold * price_last
print(-cost_money)
```
