```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot([1,2,3,4], [5,6,7,8])
plt.show()
```

## 绘制一个4*4的画布，放置4个子图

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1) # 将画布分成2*2，ax1占在第1个
ax1.plot([1,2,3,4], [5,6,7,8])
ax2 = fig.add_subplot(2,2,2) # ax2占在第2个
ax2.plot([1,2,3,4], [5,3,1,8])
ax3 = fig.add_subplot(2,2,3) # ax3占在第3个
ax3.plot([1,2,3,4], [5,1,1,4])
ax4 = fig.add_subplot(2,2,4) # ax4占在第4个
ax4.plot([1,2,3,4], [5,-2,1,4])
plt.show()
```
