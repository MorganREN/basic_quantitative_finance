```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4], [2,4,1,7], 'D-y', label='Line A') # 折线图
plt.plot([1,2,3,4], [2,2,5,3], 'D-g', label='Line B')
plt.title('Matplotlib Test Plot')
plt.xlabel("Xlabel")
plt.ylabel("Ylabel")
# plt.xlim(0, 5) # 范围
# plt.ylim(0, 6)
# plt.xticks(np.arange(0, 10, 2), ['a', 'b', 'c', 'd', 'e']) # 刻度
plt.legend() # 使上面两个plot的label显示出来

plt.show()
```
