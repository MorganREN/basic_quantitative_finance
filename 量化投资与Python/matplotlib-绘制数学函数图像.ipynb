## Matplotlib-绘制数学函数图象
- 使用Matplotlib模块在一个窗口中绘制数学函数y=x， y=x^2, y=3x^3+5x^2+2x+1的图象，是用不同颜色的线加以区别，并使用图例说明各个线代表什么函数。


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.linspace(-100, 100, 10000)
print(x)
```

```python
y1 = x.copy()
y2 = x ** 2
y3 = 3*x**3 + 5*x**2 + 2*x + 1
```

```python
plt.plot(x, y1, 'o-r', label='y=x')
plt.plot(x, y2, 'o-b', label='y=x^2')
plt.plot(x, y3, 'o-g', label='y=3x^3+5x^2+2x+1')
plt.legend()
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```
