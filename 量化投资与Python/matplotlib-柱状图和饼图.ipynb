```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 条形图

```python
plt.bar([0,1,2,3], [5,6,7,8])
plt.show()
```

```python
plt.bar([0,1,2,4], [5,6,7,8]) # 第一个参数为位置
plt.show()
```

```python
data = [32,41,52,50]
labels = ['Jan', 'Feb', 'Mar', 'Apr']

plt.bar(np.arange(len(data)), data, color=['red', 'green','blue','yellow'], width=0.4,align='edge')
plt.xticks(np.arange(len(data)), labels)
plt.show()
```

## 饼图

```python
plt.pie([10,20,30,40])
plt.show()
```

```python
plt.pie([10,20,30,40],labels=['a','b','c','d']) # 加labels
plt.show()
```

```python
plt.pie([10,20,30,40],labels=['a','b','c','d'],autopct='%.0f%%') # 加百分比
plt.show()
```

```python
plt.pie([10,20,30,40],labels=['a','b','c','d'],autopct='%.0f%%',explode=[0.1,0.2,0,0]) # 突出某部分（数值代表突出的百分比）
plt.show()
```
